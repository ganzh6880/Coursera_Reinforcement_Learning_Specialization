## 2 - 1000-State RandomWalk Environment

First, let's create the environment. Here, we will implement an environment with a large number of states. When the state space is large, it can be problematic to store a large table of memory, while also being constrained by time to fill all the entries correctly. By using generalization, we can significantly speed learning as we will show through this assignment. We will implement the 1000-state random walk task (Example 9.1 (p.203) in Sutton and Barto). 


![](data/randomwalk_diagram.png)


### Environment Description


There are 1000 states numbered from 1 to 1000, left to right, and all episodes begin with the agent located at the center, in state 500.

The episode terminates when the agent goes far left beyond state 1 or far right beyond state 1000. 
For implementation simplicity, we will consider state 0 and state 1001 as the left/right terminal state respectively. Termination on the left (state 0) gives the agent a reward of -1, and termination on the right (state 1001) gives the agent a reward of +1.

Description of the Agent’s fixed policy and actions is given later when we implement the Agent, and for now you can regard Agent's action passed to the Environment as the amount of state to move left or right.

## 2-1: Implement Environment Methods

In this section, you will implement `env_init()`, `env_start()`, and `env_step()`.

```python
class RandomWalkEnvironment(BaseEnvironment):
    def env_init(self, env_info={}):
        """
        Setup for the environment called when the experiment first starts.
        
        Set parameters needed to setup the 1000-state random walk environment.
        
        Assume env_info dict contains:
        {
            num_states: 1000,
            start_state: 500,
            left_terminal_state: 0,
            right_terminal_state: 1001 
        }
        """
        ### START CODE HERE ### 4 lines
        
        self.num_states = env_info["num_states"] 
        self.start_state = env_info["start_state"] 
        self.left_terminal_state = env_info["left_terminal_state"] 
        self.right_terminal_state = env_info["right_terminal_state"]
        
        ### END CODE HERE ###

    def env_start(self):
        """
        The first method called when the experiment starts, called before the
        agent starts.

        Returns:
            The first state observation from the environment.
        """

        ### START CODE HERE ### 1 line
        
        # set self.reward_obs_term tuple accordingly
        self.reward_obs_term = (0.0, self.start_state, False)
        
        ### END CODE HERE ###
        
        # return first state observation from the environment
        return self.reward_obs_term[1]
        
    def env_step(self, action):
        """A step taken by the environment.

        Args:
            action: The action taken by the agent

        Returns:
            (float, state, Boolean): a tuple of the reward, state observation,
                and boolean indicating if it's terminal.
        """
        
        # set last_state and current_state
        # last state: state before taking action
        # current state: state after taking action
        ### START CODE HERE ### 2 lines

        last_state = self.reward_obs_term[1]
        current_state = last_state + action  # action represents how many states to move from current state

        ### END CODE HERE ###
        
        # set reward and is_terminal
        ### START CODE HERE ### 8~10 lines
        
        reward = 0.0
        is_terminal = False
        if current_state == self.left_terminal_state: 
            reward = -1.0
            is_terminal = True

        elif current_state == self.right_terminal_state:
            reward = 1.0
            is_terminal = True

        ### END CODE HERE ###
        
        self.reward_obs_term = (reward, current_state, is_terminal)
        
        return self.reward_obs_term
        
```

## 3 - semi-gradient TD(0) Agent with State Aggregation

Now that you have implemented the Environment, let's create the Agent that interacts with the Environment.

You will create an Agent that learns with semi-gradient TD(0) with state aggregation. 

For state aggregation, if the resolution (num. of aggregated states) is 10, then 1000 states are partitioned into 10 groups of 100 states each (i.e., states 1-100 are one group, states 101-200 are another, and so on.) 

Hence, 100 states would share the same feature, and there would be 10 distinct features (one-hot vector of size 10)

### Agent Fixed Policy Description

For this assignment, we will be performing **policy evaluation**, estimating state-value function of a fixed policy. The agent’s fixed policy is as follows:

From its current state, the agent transitions either to one of the 100 neighboring states on its left or to one of the 100 neighboring states on its right, with equal probability. (The agent has to move to a different state from its current state.)

When the agent is near the edge, then there may be fewer than 100 neighboring states on that side. In that case, all probability that would have gone into those missing neighbors goes into the probability of terminating on that side (Thus, the agent at state 1 has a 0.5 chance of terminating on the left, and the agent at state 950 has a 0.25 chance of terminating on the right).

You can regard agent's actions to be the amount of state to move left or right. (e.g. If the agent is at state 1, it will output action -1 with a probability of 0.5, and one of the actions [1,2,...,100] with a probability of 0.5)


In this section, you will be implementing the following methods below:

```python
# Do not modify this cell!

# Create empty TDAgent class
# We will fill these methods later
class TDAgent(BaseAgent):
    def __init__(self):
        self.name = "td_agent" # later for keeping track of different agents
        self.num_states = None
        self.num_aggregated_states = None
        self.alpha = None
        self.gamma = None
        
    def agent_init(self, agent_info={}):
        raise NotImplementedError

    def agent_start(self, observation):
        raise NotImplementedError

    def agent_step(self, reward, observation):
        raise NotImplementedError

    def agent_end(self, reward):
        raise NotImplementedError
        
    # helper method
    def agent_policy(self, observation):
        raise NotImplementedError
    
    # helper method
    def agent_get_state_feature(self, observation):
        raise NotImplementedError
        
    def agent_message(self, message):
        raise NotImplementedError
```

## 3-1: Implement Useful Functions

Before we start creating the Agent, we will need to define a couple functions that would be useful for the Agent.

## 3-1a: Selecting Actions 

In this section, you will implement `agent_policy()`.

Given observation (from the environment), this method should output actions according to the agent's fixed policy. 

`agent_policy()` will be later used in `agent_start()` and `agent_step()` when the agent selects actions.

Refer back to Section 2 above for description of transition.

```python
%%add_to TDAgent
def agent_policy(self, observation):
    possible_actions = []
    ### START CODE HERE ### 5~10 lines
    # add action to possible actions list, proportional to the probability of taking that action 
    # (i.e. there would be multiple duplicate entries if that action has high probability of being taken)

    for action in range(-100, 100 + 1):
        if action < 0:
            possible_actions.append(max(-observation, action))
        elif action == 0:
            pass
        elif action > 0:
            possible_actions.append(min(self.num_states + 1 - observation, action))

    ### END CODE HERE ###

    # choose from possible_transitions with equal probability
    chosen_action = self.rand_generator.choice(possible_actions)
    return chosen_action
    
```

## 3-1b: Processing State Features with State Aggregation

In this section you will implement `agent_get_state_feature()`

This method takes in an observation and returns the aggregated state feature (one-hot-vector) of that observation.

`agent_get_state_feature()` will be used in `agent_step()` and `agent_end()` when the agent updates its state values.

```python
%%add_to TDAgent
def agent_get_state_feature(self, observation):
    """
    Given observation, return the feature of that observation
    """
    # Create one-hot-vector with size of the num_aggregated_states
    # if num_aggregated_states = 10, create one-hot-vector of size 10
    
    ### START CODE HERE ### 3~4 lines
    
    one_hot_vector = np.zeros(self.num_aggregated_states)
    num_states_in_group = int(self.num_states / self.num_aggregated_states)

    one_hot_vector[int((observation - 1) / num_states_in_group)] = 1

    ### END CODE HERE ###
    return one_hot_vector
```

## 3-2: Implement Agent Methods

Now that we have implemented all the useful functions for the Agent, let's create the Agent.

In this section, you will implement `agent_init()`, `agent_start()`, `agent_step()` and `agent_end()`.

You will have to use `agent_policy()` and `agent_get_state_feature()` that you implemented above.


When saving state values in the agent, recall how the state values are represented with linear function approximation.

**State Value Representation**: $\hat{v}(s,\mathbf{w}) = \mathbf{w}\cdot\mathbf{x^T}$ where $\mathbf{w}$ is a weight vector and $\mathbf{x}$ is the feature vector of the state.



When performing TD(0) updates with Linear Function Approximation, recall how we perform semi-gradient TD(0) updates using supervised learning.

**semi-gradient TD(0) Weight Update Rule**: $\mathbf{w_{t+1}} = \mathbf{w_{t}} + \alpha [R_{t+1} + \gamma \hat{v}(S_{t+1},\mathbf{w}) - \hat{v}(S_t,\mathbf{w})] \nabla \hat{v}(S_t,\mathbf{w})$

```python
%%add_to TDAgent
def agent_init(self, agent_info={}):
    """Setup for the agent called when the experiment first starts.

    Set parameters needed to setup the semi-gradient TD(0) state aggregation agent.

    Assume agent_info dict contains:
    {
        num_states: integer, # 1000
        num_aggregated_states: integer, # 10
        alpha: float, # learning rate (0.1)
        gamma: float, # discount factor (1.0) 
    }
    """

    self.rand_generator = np.random.RandomState(agent_info.get("seed")) # set random seed for each run
    
    # save relevant info from agent_info
    ### START CODE HERE ### 4 lines
    
    self.num_states = agent_info.get("num_states")
    self.num_aggregated_states = agent_info.get("num_aggregated_states")
    self.alpha = agent_info.get("alpha")
    self.gamma = agent_info.get("gamma")

    ### END CODE HERE ###
    
    
    # initialize weights
    # initialize all weights to zero using numpy array
    ### START CODE HERE ### 1 line
    
    self.weights = np.zeros(self.num_aggregated_states)
    
    ### END CODE HERE ###

    self.last_state = None
    self.last_action = None


def agent_start(self, observation):
    """The first method called when the experiment starts, called after
    the environment starts.
    Args:
        observation (Numpy array): the state observation from the
            environment's evn_start function.
    Returns:
        The first action the agent takes.
    """
    
    # take action given observation, using self.agent_policy()
    # save self.last_state and self.last_action to perform updates later
    ### START CODE HERE ### 2 line
    
    self.last_state = observation
    self.last_action = self.agent_policy(observation)

    ### END CODE HERE ###
    
    return self.last_action


def agent_step(self, reward, observation):
    """A step taken by the agent.
    Args:
        reward (float): the reward received for taking the last action taken
        observation (Numpy array): the state observation from the
            environment's step based, where the agent ended up after the
            last step
    Returns:
        The action the agent is taking.
    """
    # update weights and take action
    ### START CODE HERE ### 5~7 lines

    # get feature
    current_state_feature = self.agent_get_state_feature(observation)
    last_state_feature = self.agent_get_state_feature(self.last_state)

    self.weights = self.weights + self.alpha * (reward + self.gamma * np.dot(self.weights, current_state_feature.T) - np.dot(self.weights, last_state_feature.T)) * last_state_feature

    self.last_state = observation
    self.last_action = self.agent_policy(observation)

    ### END CODE HERE ###
    return self.last_action


def agent_end(self, reward):
    """Run when the agent terminates.
    Args:
        reward (float): the reward the agent received for entering the
            terminal state.
    """
    
    # update weights
    ### START CODE HERE ### 2~4 lines
    
    last_state_feature = self.agent_get_state_feature(self.last_state)

    # update weights
    self.weights = self.weights + self.alpha * (reward + self.gamma * 0.0 - np.dot(self.weights, last_state_feature.T)) * last_state_feature
    
    ### END CODE HERE ###
    return
```

**Hint**: Think about how state values are represented with linear function approximation. `state_value` array will be a 1D array with length equal to the number of original states.

```python
%%add_to TDAgent
def agent_message(self, message):
    if message == 'get state value':
        ### START CODE HERE ### 2~4 lines
        
        features = [self.agent_get_state_feature(state) for state in range(1, self.num_states + 1)]
        state_value = np.dot(self.weights, np.asarray(features).T) 
        
        ### END CODE HERE ###
        return state_value
    
```

## 4-1 : Prediction Objective (Root Mean Squared Error) 



Recall that the Prediction Objective in function approximation is Mean Squared Value Error $\overline{VE}(\mathbf{w}) \doteq \sum\limits_{s \in \mathcal{S}}\mu(s)[v_\pi(s)-\hat{v}(s,\mathbf{w})]^2$

However in practice, it is hard to compute the state distribution $\mu(s)$ in the Mean Squared Value Error, and we use Root Mean Squared Error as an approximate measure, where states are equally weighted.

Root Mean Squared Error $RMSE(\mathbf{w}) \doteq \sqrt{\frac{1}{n} \sum\limits_{s \in \mathcal{S}}[v_\pi(s)-\hat{v}(s,\mathbf{w})]^2}$ where $n$ is the total number of states.

`calc_RMSE()` computes RMSE given true state value $v_\pi(s)$ and learned state value $\hat{v}(s, \mathbf{w})$.


```python
# Do not modify this cell!

def calc_RMSE(true_state_val, learned_state_val, num_states):
    
    assert(len(true_state_val) == len(learned_state_val) == num_states)
    SSE = np.sum(np.square(true_state_val - learned_state_val))
    RMSE = np.sqrt( 1.0/num_states * SSE)

    return RMSE
```

## 4-2 : Run Experiment with 10-State Aggregation

We will first test our implementation using state aggregation with resolution of 10.

We have already provided you the experiment/plot code, so you can go ahead and run the two cells below.

Note that running the two experiment cells below will take **_approximately 3 min_**.

```python
# Do not modify this cell!

# Define function to run experiment
def run_experiment(environment, agent, environment_parameters, agent_parameters, experiment_parameters):

    # We provide you with the true state value
    true_V = np.load('data/true_V.npy')
    
    # generate random seed for runs
    np.random.seed(0) # fixed so the random_seed_arr is also fixed
    random_seed_arr = np.random.randint(10000, size=experiment_parameters["num_runs"])
    

    rl_glue = RLGlue(environment, agent)
    
    # Sweep Agent parameters
    for num_agg_states in agent_parameters["num_aggregated_states"]:
        for lr in agent_parameters["learning_rate"]:
            
            # save rmse at the end of each episode
            agent_rmse = np.zeros(experiment_parameters["num_episodes"])
            # save learned state value at the end of each run
            agent_state_val = np.zeros(environment_parameters["num_states"])

            env_info = {"num_states": environment_parameters["num_states"],
                        "start_state": environment_parameters["start_state"],
                        "left_terminal_state": environment_parameters["left_terminal_state"],
                        "right_terminal_state": environment_parameters["right_terminal_state"]}

            agent_info = {"num_states": environment_parameters["num_states"],
                          "num_aggregated_states": num_agg_states,
                          "alpha": lr,
                          "gamma": environment_parameters["discount_factor"]}

            print('Setting - num. agg. states: {}, learning rate: {}'.format(num_agg_states, lr))
            os.system('sleep 1')
            
            # one agent setting
            for run in tqdm(range(experiment_parameters["num_runs"])):
                agent_info["seed"] = random_seed_arr[run]
                rl_glue.rl_init(agent_info, env_info)

                for episode in range(experiment_parameters["num_episodes"]):
                    # run episode
                    rl_glue.rl_episode(0) # no step limit
                    
                    current_V = rl_glue.rl_agent_message("get state value")
                    agent_rmse[episode] += calc_RMSE(true_V, current_V, environment_parameters["num_states"])
                    
                agent_state_val += current_V
                # agent_msve += calc_msve(true_V, current_V)
                
            # average state value and rmse over runs
            agent_state_val /= experiment_parameters["num_runs"]
            agent_rmse /= experiment_parameters["num_runs"]
            
            save_name = "{}_agg_states_{}_lr_{}".format(rl_glue.agent.name, num_agg_states, lr).replace('.','')

            # save avg. state value
            np.save("result/V_{}".format(save_name), agent_state_val)

            # save avg. rmse
            np.save("result/RMSE_{}".format(save_name), agent_rmse)
```

```python
# Run Experiment

# Experiment parameters
experiment_parameters = {
    "num_runs" : 10,
    "num_episodes" : 1000
}

# Environment parameters
environment_parameters = {
    "num_states" : 1000,
    "start_state" : 500,
    "left_terminal_state" : 0,
    "right_terminal_state" : 1001,
    "discount_factor" : 1.0
}

# Agent parameters
# Each element is an array because we will be later sweeping over multiple values
agent_parameters = {
    "num_aggregated_states": [10],
    "learning_rate": [0.1]
}

current_env = RandomWalkEnvironment
current_agent = TDAgent

# run experiment
run_experiment(current_env, current_agent, environment_parameters, agent_parameters, experiment_parameters)

# plot result
plot_script.plot_result(agent_parameters)
```
